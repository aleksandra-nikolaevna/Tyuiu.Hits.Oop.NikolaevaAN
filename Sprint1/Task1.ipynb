```csharp
public class Animal
{
    private string _color;
    private double _age;

    // Свойство Color с геттером и сеттером
    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    // Свойство Age с геттером и сеттером
    public double Age
    {
        get { return _age; }
        set
        {
            if (value > 0)
                _age = value;
            else
                throw new ArgumentOutOfRangeException("Возраст не может быть отрицательным или равным 0!");
        }
    }

    // Статический атрибут
    public static int Satiety { get; private set; } = 0;

    // Конструктор
    public Animal()
    {
        Satiety++;
    }

    // Обычный метод
    public void Run(int delta) 
    {
        Speed += delta; 
    }

    // Статический метод
    public static void DisplayTotalAnimals()
    {
        Console.WriteLine($"Общее количество животных в зоопарке: {TotalAnimals}");
    }
}

public class Penguin : Animal
{
    private double _height;

    // Свойство Height с геттером и сеттером
    public double Height
    {
        get { return _height; }
        set { _height = value; }
    }

    // Конструктор
    public Penguin() : base() 
    {
        Height = 0.6; // Задаем рост животного
    }

    // Обычный метод
    public void GetHigher(double amount) 
    {
        Height += amount;
    }
}

public class NavySeal : Animal
{
    private double _height;

    // Свойство Height с геттером и сеттером
    public double Height
    {
        get { return _height; }
        set { _height = value; }
    }

    // Конструктор
    public NavySeal() : base() 
    {
        Height = 1.5; // Задаем рост животного
    }

    // Обычный метод
    public void GetHigher(double amount) 
    {
        Height += amount;
    }
}
```
