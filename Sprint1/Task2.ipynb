```csharp
public class Animal
{
    public string color // public свойство
    private int age; // private поле

    public string Color
    {
        get { return color; }
        set { color = value; }
    }

    protected void SetAge(int value) // protected метод
    {
        age = value;
    }
}

public class Pig : Animal
{
    private int weight; // private поле

    public void Feed(double amount) // public метод
    {
        weight += amount;
    }

    protected void CheckWeight() // protected метод
    {
        if (weight > 130)
        {
            Console.WriteLine("Достигнут оптимальный вес");
        }
    }
}

public class Elephant : Animal
{
    private int drunkWater; // private поле

    public void DrunkWater(double amount) // public метод
    {
        drunkWater += amount;
    }

    protected void CheckDrunkWater() // protected метод
    {
        if (drunkWater >= 250)
        {
            Console.WriteLine("Возможно заболевание у слона");
        }
    }
}

Animal myAnimal = new Animal();
Pig myPig = new Pig();
Elephant myElephant = new Elephant();

myAnimal.Color = "White";
myAnimal.SetAge(6);

myPig.Color = "Pink";
myPig.Feed(0.5); // Увеличиваем вес свинки

myElephant.Color = "Gray";
myElephant.DrunkWater(2.9); // Увеличиваем кол-во выпитой воды
```
